```python
def print_matriks(matrik):
    for row in matrik:
        print(row)
    print()
    
def pertambahan_matriks(A,B):
    return [
        [A[0][0] + B[0][0], A[0][1] + B[0][1]],
        [A[1][0] + B[1][0], A[1][1] + B[1][1]]
    ]

def pengurangan_matriks(A,B) : 
    return [
        [A[0][0] - B[0][0], A[0][1] - B[0][1]],
        [A[1][0] - B[1][0], A[1][1] - B[1][1]]
    ]

def perkalian_matriks(A,B) : 
    return [
        [A[0][0] * B[0][0] + A[0][1] * A[1][0], A[0][0] * B[0][1] + A[0][1] * B[1][1]],
        [A[1][0] * B[0][0] + A[1][1] * A[1][0], A[1][0] * B[0][1] + A[1][1] * B[1][1]]
    ]

matriksA = [[4,5],
            [6,3]]

matriksB = [[6,2], 
            [5,3]]

def main():
    print(f'Matriks A : ')
    print_matriks(matriksA)
    print(f'Matriks B :')
    print_matriks(matriksB)

    while True : 
        operasi = int(input('Pilih operasi apa : 1. Pertambahan 2. Pengurangan 3. Perkalian 4. Keluar'))
        if operasi == 1 : 
            print(f'Hasil dari Matriks A + Matriks B : ')
            print_matriks(pertambahan_matriks(matriksA,matriksB))   
            print()
        elif operasi == 2 :
            print(f'Hasil dari Matriks A - Matriks B : ')
            print_matriks(pengurangan_matriks(matriksA, matriksB))
            print()
        elif operasi == 3:
            print('Hasil dari Matriks A * Matriks B : ')
            print_matriks(perkalian_matriks(matriksA, matriksB))
            print()
        elif operasi == 4:
            print('Anda memilih keluar terima kasih !')
            break
        else :
            break

main()
```

```python
def input_set():
    data_set = set()
    while True: 
        data = input('Masukkan data yang ingin di input : (ketik "stop" untuk berhenti): ')
        if data.lower() == 'stop':
            return data_set
        data_set.add(data)

def input_tuple():
    data_tuple = []
    while True:
        data = input('Masukkan data yang ingin di input : (ketik "stop" untuk berhenti): ')
        if data.lower() == 'stop':
            return tuple(data_tuple)
        data_tuple.append(data)

while True:
    tipe_data = input('Tipe data apa yang ingin di buat (string/tuple/set)? ("cancel" atau "stop" untuk keluar!): ')
    if tipe_data.lower() == 'string':
        data_string = input('Silahkan masukkan string yang ingin anda buat: ')
        print(f'String anda adalah: "{data_string}"')
    elif tipe_data.lower() == 'set':
        data = input_set()
        print(f'Data set yang anda buat adalah: {data}')
    elif tipe_data.lower() == 'tuple':
        data = input_tuple()
        print(f'Data tuple yang anda buat adalah: {data}')
    elif tipe_data.lower() in ['cancel', 'stop']:
        print('Anda memilih keluar. Terima kasih!!!')
        break
    else:
        print('Kesalahan! : Pastikan anda mengetikkan tipe data yang benar!')
```

```python
#Buatlah program dengan tipe data dictionaries dengan menu pilihan menu input data, cetak data, hitung jumlah record dictionary, update, delete data !

def add_data(profile_mahasiswa):
    key_records = int(input('Berapa banyak key yang ingin anda tambahkan? : '))
    for i in range(key_records) :
        key = input('Silahkan tambahkan key untuk data : ')
        value_records = int(input(f'Berapa banyak value yang ingin anda tambahkan untuk key : {key} ? : '))
        list_value = []
        for j in range(value_records):
            value = input('Silahkan masukkan Value nya : ')
            if value_records == 1 :
                profile_mahasiswa[key] = value
                break
            else:
                list_value.append(value)
            profile_mahasiswa[key] = list_value    
    print('Data berhasil di tambahkan!')
    return profile_mahasiswa
    
    
def print_data(profile_mahasiswa):
    print('Mencetak dictionary : ')
    if len(profile_mahasiswa) == 0:
        print('Data masih kosong!')
    else:
        print(profile_mahasiswa)
        for key,value in profile_mahasiswa.items() :
            print(f'{key} : {value}')
        print('Semua data telah di cetak!')
    
def length_data(profile_mahasiswa):
    length_data = len(profile_mahasiswa)
    print(f'Record dictionary dari data sebanyak {length_data} data.')
    return length_data
    
def update_data(profile_mahasiswa):
    key = input('Silahkan masukkan key dari data yang ingin di update : ')
    if key in profile_mahasiswa.keys():
        value_records = int(input('Berapa banyak data yang ingin di tambahkan : '))
        list_value = []
        for i in range(value_records) :
            value = input('Masukkan data yang ingin di tambahkan : ')
            if value_records == 1:
                profile_mahasiswa.update({f'{key}' : f'{value}'})     
                break
            else:
                list_value.append(value)
            profile_mahasiswa.update({f'{key}' : f'{list_value}'})        
    else:
        print('Key tidak di temukan!')
    print('Data berhasil di update!')
    return profile_mahasiswa
    
def delete_data(profile_mahasiswa):
    key = input('Silahkan masukkan key dari data yang ingin di update : ')
    if key in profile_mahasiswa.keys():
        print(f'{key} Ditemukan !!!')
        print(f'Apakah anda yakin ingin menghapus "{key}" : "{profile_mahasiswa[key]}"?')
        confirm = input('y/n').lower()
        if confirm == 'y':
            print(f'Data {key} : {profile_mahasiswa[key]} berhasil di hapus! ')
            del profile_mahasiswa[key]
        else:
            print('Penghapusan di batalkan')
        return profile_mahasiswa
            


def main(profile_mahasiswa):
    while True :
        operation = int(input('Silahkan pilih operasi : 1. Tambah Data 2. Cetak Data 3.Hitung Jumlah Data 4. Update Data 5. Delete Data 6. Keluar : '))
        if operation == 1 :
            add_data(profile_mahasiswa)
            print()
        elif operation == 2 :
            print_data(profile_mahasiswa)
            print()
        elif operation == 3:
            length_data(profile_mahasiswa)
            print()
        elif operation == 4:
            update_data(profile_mahasiswa)
            print()
        elif operation == 5:
            delete_data(profile_mahasiswa)
            print()
        elif operation == 6:
            print('Anda memilih keluar. Terima Kasih!!!')
            break
        else:
            print('Pilihan anda salah : Silahkan masukkan pilihan yang benar!!!')
            print()

profile_mahasiswa = {}
main(profile_mahasiswa)
```
