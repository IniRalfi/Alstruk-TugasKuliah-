```python
# Python program for implementation of Insertion Sort

# Function to sort array using insertion sort
def insertion_sort(data):
    for i in range(1, (len(data))):
        # data = [64,10,12,15,8] 
        key = data[i]
        # key = data[1] = 10
        j = i - 1
        # j = 1 - 1 = 0
        # Geser elemen yang lebih besar dari key ke kanan
        while j >= 0 and key < data[j]:
            # j = 0 dan 10 < data[0]
            # j =0 dan 10 < 64 : True
            data[j + 1] = data[j]
            # data [ 0 + 1] = data [0]
            # data[1] = 64
            j -= 1
            # j = 0 -1 = -1
            # data = [64,64,12,15,8]
            
        #while j >= 0 and data[j].....
        # j = -1 , j < 0 (False) keluar While
        data[j + 1] = key
        # data [-1 + 1] = 10
        #data[0] = 10
        # arr setelah iterasi pertama: [10,64,12,15,8]

        # Iterasi i = 2
        key = data[i]
        #data = [10,64,12,15,8]
        # key = data[2] = 12
        j = i - 1
        # j = 2 - 1 = 1
        while j >= 0 and key < data[j]:
            # j = 1 dan 12 < data[1]
            # j = 1 True dan 12 < 64 True : 
            data[j + 1] = data[j]
            # data[1+1] = data[1]
            #data [2] = 64
            
            #data = [10, 64, 64, 15, 8]
            j -= 1
            # j = 1 - 1 = 0
            
        # while j >= 0 and key < data[j] : 
        # j = 0 True
        # 12 < 10 False
        #Keluar while
        data[j + 1] = key
        # data [0 + 1]  = 12
        # data[1] = 12
        # arr setelah iterasi kedua: [10, 12, 64, 15, 8]

        # Iterasi i = 3
        key = data[i]
        # data = [10, 12, 64, 15, 8]
        # key = arr[3] = 15
        j = i - 1
        # j = 3 - 1 = 2
        while j >= 0 and key < data[j]:
        # j = 2 True dan 15 < data[2] == 15 < 64 True: 
            data[j + 1] = data[j]
            # data [ 2+1] = data[2]
            # data[3] = 64
            
            #data = [10,12,64,64,8]
            j -= 1
            # j = 2 -1 = 1
            
        #  while 1 >= 0 and 15 < data[1] == 15 < 12 False
        
        data[j + 1] = key
        # data [1 + 1] = 15
        #data [2] = 15
        # data setelah iterasi ketiga: [10, 12, 15, 64, 8]

        # Iterasi i = 4
        key = data[i]
        # key = data[4] = 8
        j = i - 1
        # j = 4 - 1 = 3
        while j >= 0 and key < data[j]:
        # while 3 >= 0 and 8 < 15 False
            data[j + 1] = data[j]
            # data[4] = 64
            # data = [10, 12, 15, 15, 64]
            j -= 1
            # j = 3 -1 = 2
            
        #while 2 >= 0 and 8 < data[2] == 8 < 15
            data[j + 1] = data[j]
            # data[3] = 15
            # data = [10, 12, 12, 15, 64]
            j -= 1
            # j = 2 -1 = 1
        
        #while 1 >= 0 and 8 < data[1] == 8 < 12 True : 
            data[j + 1] = data[j]
            # data[2] = 12
            # data = [10, 10, 12, 15, 64]
            j -= 1
            # j = 1 -1 = 0
            
        #while 0 >= 0 and 8 < data[0] == 8 < 10 True :
            data[j + 1] = data[j]
            # data[1] = 10
            # data = [10, 10, 12, 15, 64]
            j -= 1
            # j = 0 -1 = -1
            
        #while -1 >= 0 and 8 < data[0] == 8 < 10 False :
        
        data[j + 1] = key
        # data [0] = 8
        # arr setelah iterasi keempat: [8, 10, 12, 15, 64]
        return data
    
    data = [64,10,12,15,8]    
    print(insertion_sort(data)) 
    print("Sorted array: ", end="")
```
